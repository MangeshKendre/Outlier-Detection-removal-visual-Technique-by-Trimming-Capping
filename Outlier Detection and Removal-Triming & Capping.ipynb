# Outlier removal techniques

# 1. Using Inter quartile Range (IQR) with applying Trimming method


```python
# Dictionary to store outlier information
outliers = {}

# New DataFrame to store the trimmed data
new_df = df.copy()

# Loop through the columns
for i in num_var:
    
     print(i)
     print()  
     print()
    
     # Calculate the IQR
     q1 = df[i].quantile(0.25)
     q3 = df[i].quantile(0.75)
     iqr = q3 - q1
     print(f"Column: '{i}'")
     print(f"Before trimming - Shape: {df[i].shape}")
     print(f"q1: {q1}")
     print(f"q3: {q3}")
     print(f"iqr: {iqr}")

     # Find UpperLimit and LowerLimit
     upper_limit = q3 + 1.5 * iqr
     lower_limit = q1 - 1.5 * iqr
     print(f"Upper_limit: {upper_limit}")
     print(f"Lower_limit: {lower_limit}")
    
    
     # Find the outliers
     outlier_indices = df[(df[i] < lower_limit) | (df[i] > upper_limit)].index
 
    # Store the outlier indices in the dictionary
     outliers[i] = outlier_indices
        
     # Trim the outliers by removing them from the new DataFrame
     new_df = new_df[~new_df.index.isin(outlier_indices)]
     print(f"After trimming - Shape: {new_df[i].shape}")
     print()
     print()  
    
    #visualize the trimming graphs
    
    # Create subplots
     fig, axes = plt.subplots(2, 2, figsize=(12, 6))
        
     # Box plot
     sns.boxplot(x=df[i], ax=axes[0, 0])
     axes[0, 0].set_xlabel(i)
     axes[0, 0].set_title('Box Plot - Before Trimming')
        
     # Distribution plot (distplot)
     sns.distplot(df[i].dropna(), ax=axes[0, 1])
     axes[0, 1].set_xlabel(i)
     axes[0, 1].set_ylabel('Density')
     axes[0, 1].set_title('Distribution Plot - Before Trimming')
    
     # Box plot with updated data
     sns.boxplot(x=new_df[i], ax=axes[1, 0])
     axes[1, 0].set_xlabel(i)
     axes[1, 0].set_title('Box Plot - After Trimming')
    
     # Distribution plot (distplot) with updated data
     sns.distplot(new_df[i].dropna(), ax=axes[1, 1])
     axes[1, 1].set_xlabel(i)
     axes[1, 1].set_ylabel('Density')
     axes[1, 1].set_title('Distribution Plot - After Trimming')
        
     # Adjust layout
     plt.tight_layout()
        
     # Show the plots
     plt.show()

        
# Print the shape of the trimmed DataFrame
print(f"Shape of the trimmed DataFrame: {new_df.shape}")
    
    
    
    
    
    
```

# 2. Using Inter quartile Range (IQR) with applying Capping method


```python
# Dictionary to store outlier information
outliers = {}

# New DataFrame to store the trimmed data
new_df_cap = df.copy()

# Loop through the columns
for i in num_var:
     print(i)
     print()
     # Calculate the IQR
     q1 = df[i].quantile(0.25)
     q3 = df[i].quantile(0.75)
     iqr = q3 - q1
     print(f"Column: '{i}'")
     print(f"Before capping - Shape: {df[i].shape}")
     print(f"q1: {q1}")
     print(f"q3: {q3}")
     print(f"iqr: {iqr}")

     # Find UpperLimit and LowerLimit
     upper_limit = q3 + 1.5 * iqr
     lower_limit = q1 - 1.5 * iqr
     print(f"Upper_limit: {upper_limit}")
     print(f"Lower_limit: {lower_limit}")
    
     # Find the outliers
     outlier_indices = df[(df[i] < lower_limit) | (df[i] > upper_limit)].index
     
     # Store the outlier indices in the dictionary
     outliers[i] = outlier_indices
        
     # Trim the outliers by capping them in the new DataFrame
     new_df_cap[i] = np.where(
     new_df_cap[i] > upper_limit,
     upper_limit,
     np.where(
     new_df_cap[i] < lower_limit,
     lower_limit,
     new_df_cap[i]
 )
 )
     print(f"After capping - Shape: {new_df_cap[i].shape}")
     print()
        
     # Create subplots
     fig, axes = plt.subplots(2, 2, figsize=(12, 6))
        
     # Box plot
     sns.boxplot(x=df[i], ax=axes[0, 0])
     axes[0, 0].set_xlabel(i)
     axes[0, 0].set_title('Box Plot - Before capping')
        
     # Distribution plot (distplot)
     sns.distplot(df[i].dropna(), ax=axes[0, 1])
     axes[0, 1].set_xlabel(i)
     axes[0, 1].set_ylabel('Density')
     axes[0, 1].set_title('Distribution Plot - Before capping')
    
     # Box plot with updated data
     sns.boxplot(x=new_df_cap[i], ax=axes[1, 0])
     axes[1, 0].set_xlabel(i)
     axes[1, 0].set_title('Box Plot - After capping')
    
     # Distribution plot (distplot) with updated data
     sns.distplot(new_df_cap[i].dropna(), ax=axes[1, 1])
     axes[1, 1].set_xlabel(i)
     axes[1, 1].set_ylabel('Density')
     axes[1, 1].set_title('Distribution Plot - After capping')
        
     # Adjust layout
     plt.tight_layout()
        
     # Show the plots
     plt.show()
   
        
        
        
# Print the shape of the Capping DataFrame
print(f"Shape of the Capping DataFrame: {new_df_cap.shape}")
```
